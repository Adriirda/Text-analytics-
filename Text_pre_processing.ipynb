```python
# Import necessary libraries
import numpy as np
import pandas as pd
import re
import nltk
import spacy
import string
```

```python
# Read data
full_df = pd.read_csv("group_chat.csv")
df = full_df[["message"]]
df["message"]= df["message"].astype(str)
full_df.head()
```

#Text processing


1.   Convert to lower case
2.   Remove URLs
3.   Remove punctuations
4.   Remove emoji
4.   Remove stopwords
6.   Tokenize word
5.   Stemming
6.   Lemmatization




```python
# FUNCTION LISTS

# Function to remove URLs
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

# Function to special characters
def remove_special_characters(text):
    return re.sub(r'[^a-zA-Z\s]', '', text)

# Function to remove emoji
def remove_emoji(string):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F700-\U0001F77F"  # alchemical symbols
                               u"\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
                               u"\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
                               u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
                               u"\U0001FA00-\U0001FA6F"  # Chess Symbols
                               u"\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
                               u"\U00002702-\U000027B0"  # Dingbats
                               u"\U000024C2-\U0001F251"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)
```

```python
# Convert the messages to lowercase
df['clean_message'] = df['message'].str.lower()

# Remove urls
df['clean_message'] = df['clean_message'].apply(remove_urls)

# Remove punctuation
df['clean_message'] = df['clean_message'].apply(remove_special_characters)

# Remove emoji
df['clean_message'] = df['clean_message'].apply(remove_emoji)
```

Tokenize words

```python
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
```

```python
# Function to tokenize
def tokenize_words(text):
    return word_tokenize(text)

df['clean_message'] = df['clean_message'].apply(tokenize_words)

# Print the DataFrame with tokenized text
df.head(20)
```

Remove stopwords

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
", ".join(stopwords.words('english'))
```

```python
stop_words = stopwords.words('english')
def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in stop_words])

df['clean_message'] = df['clean_message'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))

df.head(20)
```

Stemming

```python
from nltk.stem.porter import PorterStemmer

stemmer = PorterStemmer()
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])

df["clean_message"] = df["clean_message"].apply(lambda text: stem_words(text))
df.head(20)
```

Lemmatization

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])

df["clean_message"] = df["clean_message"].apply(lambda text: lemmatize_words(text))
df.head(20)
```

```python
nltk.download('averaged_perceptron_tagger')
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
wordnet_map = {"N":wordnet.NOUN, "V":wordnet.VERB, "J":wordnet.ADJ, "R":wordnet.ADV}
def lemmatize_words(text):
    pos_tagged_text = nltk.pos_tag(text.split())
    return " ".join([lemmatizer.lemmatize(word, wordnet_map.get(pos[0], wordnet.NOUN)) for word, pos in pos_tagged_text])

df["clean_message"] = df["clean_message"].apply(lambda text: lemmatize_words(text))
df.head(20)
```

```python
df.drop(["message"], axis=1, inplace=True)
```
